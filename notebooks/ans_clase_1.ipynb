### Introducción a Pandas

```python
df = pd.read_csv('../data/adult.csv')
df = df.drop(columns = ['Unnamed: 0'])
df.head(5)
```

```python
df.tail(5)
```

```python
### Renombrando las columnas
df.rename(columns = {'education num':'education_num',
                   'marital status':'marital_tatus',
                   'capital gain': 'capital_gain',
                   'capital loss': 'capital_loss',
                   'hours per week': 'hours_per_week',
                   'native country': 'native_country'}, 
                   inplace = True)
```

### Usando iloc

```python
### En el siguiente ejemplo se ubican las filas desde la 10 hasta la 20 y las dos primera columnas

df.iloc[10:21, :2]
```

### Usando loc

```python
### Localizar filas de acuerdo a una condición
### Obtener las filas en las que los años de educación sean iguales a 16

df.loc[df['education_num'] == 16]
```

```python
#Ejercicio: Encontrar las personas cuyo trabajo por horas sea igual o mayor a 45
df.loc[df['hours_per_week'] >= 45]
```
