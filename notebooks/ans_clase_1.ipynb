```python
def suma_dos_numeros(a, b):
    """
    Esta función toma dos números y entrega la suma de estos.

    Parameters:
        a (float): número a sumar.
        b (float): número a sumar.

    Returns:
        resultado (float): suma de a + b
    """
    resultado = a + b

    return(resultado)
```

### Creando una función

```python
def grupo_edad(age):
    if age < 25:
        return 'Young'
    elif age < 45:
        return 'Young adult'
    elif age <65:
        return 'Adult'
    else: 
        return 'Senior citizen'
```

```python
df['age_group'] = df['age'].apply(grupo_edad)
df[['age', 'age_group']].head()
```
