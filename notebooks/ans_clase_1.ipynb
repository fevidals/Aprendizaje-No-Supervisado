### Creando una función

```python
def grupo_edad(age):
    if age < 25:
        return 'Young'
    elif age < 45:
        return 'Young adult'
    elif age <65:
        return 'Adult'
    else: 
        return 'Senior citizen'
```

```python
df['age_group'] = df['age'].apply(grupo_edad)
df[['age', 'age_group']].head()
```

### Simulación de datos pseudo-aleatorios

```python
rng = np.random.RandomState(1)
```

```python
X = np.dot(rng.rand(2, 2), rng.randn(2, 200)).T
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.axis('equal');
```

### Componentes principales

```python
pca = PCA(n_components = 2)
pca.fit(X)
```

```python
print(pca.components_)
```

```python
print(pca.explained_variance_)
```
